```python
column_amid_array = full_list_df['AMID'].to_numpy()
column_am_array = full_list_df['AM'].to_numpy()

column_amid_array = column_amid_array[2:]
column_am_array = column_am_array[2:]
```

```python
no_of_values = column_amid_array.size
```

```python
calculation_df = pd.DataFrame()

calculation_df['AMID']= column_amid_array
calculation_df['AM']= column_am_array

calculation_df['difference'] = calculation_df['AMID'] - calculation_df['AM']
```

```python
calculation_df['squared_difference'] = calculation_df['difference'] * calculation_df['difference']
```

```python
sum_of_squared_deviations = calculation_df['squared_difference'].sum()
```

```python
standard_deviation = ( sum_of_squared_deviations / no_of_values) ** 0.5
```

```python
print ('AMID', 'AM', standard_deviation)
```
