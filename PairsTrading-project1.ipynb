```python
def calculate_standard_deviation(stock1='AMID', stock2='AM'):
    stock1_values = full_list_df[stock1].to_numpy()
    stock2_values = full_list_df[stock2].to_numpy()

    stock1_values = stock1_values[2:]
    stock2_values = stock2_values[2:]

    no_of_values = stock1_values.size
    
    calculation_df = pd.DataFrame()

    calculation_df[stock1]= stock1_values
    calculation_df[stock2]= stock2_values

    calculation_df['difference'] = calculation_df[stock1] - calculation_df[stock2]
    calculation_df['squared_difference'] = calculation_df['difference'] * calculation_df['difference']
    sum_of_squared_deviations = calculation_df['squared_difference'].sum()
    standard_deviation = ( sum_of_squared_deviations / no_of_values) ** 0.5
    #print ( stock1, stock2, standard_deviation)
    return standard_deviation
```

```python
stock1='AMID'

previous_standard_deviation = 10000
lowest_standard_deviation = 10000
lowest_pair = ''

loop_counter = 0
for name in Header:
    if name != stock1 and name != 'QuoteDate':
        stock2 = name
        standard_deviation = calculate_standard_deviation(stock1, stock2)
        if standard_deviation < lowest_standard_deviation:
            lowest_standard_deviation = standard_deviation
            lowest_pair = stock2

print (stock1, lowest_pair, lowest_standard_deviation)
            
```
